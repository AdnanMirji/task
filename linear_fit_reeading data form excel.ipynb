```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/Task_1/Book1.csv')

#rows are samples
#columns are features

dataset=data.iloc[:,0:1].values
print(dataset)

dataset1=data.iloc[:,1:2].values
print(dataset1)

x=np.array(dataset).reshape((-1,1))

y=np.array(dataset1).reshape((-1,1))
plt.scatter(x,y)
plt.show


from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)

pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y,pred)
plt.scatter(x,y)
plt.plot(x,pred,color='Red',marker='*')

print("MSE:",mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```
